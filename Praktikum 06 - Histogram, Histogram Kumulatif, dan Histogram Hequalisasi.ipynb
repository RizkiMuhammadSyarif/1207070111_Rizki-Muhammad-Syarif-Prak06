Import Library

```python
# memanggil library 
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

Membaca Gambar

```python
# digunakan untuk memuat sebuah file gambar bernama "Shikamaru.jpeg" 
# ke dalam sebuah variabel img menggunakan fungsi imread() dari pustaka imageio
img = imageio.imread("Shikamaru.jpeg")
```

```python
# Mendapatkan resolusi dan type dari gambar
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
```

Merubah gambar menjadi Grayscale

```python
# digunakan untuk membuat sebuah array Numpy baru yang merepresentasikan gambar grayscale 
# dengan ukuran dan tipe data yang sama dengan gambar awal (img)
img_grayscale = np.zeros(img.shape, dtype=np.uint8)

for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]
        green = img[y][x][1]
        blue = img[y][x][2]
        gray = (int(red) + int(green) + int(blue)) / 3
        img_grayscale[y][x] = (gray, gray, gray)
        
# digunakan untuk menampilkan citra grayscale yang sudah dihasilkan pada plot    
plt.imshow(img_grayscale)
plt.title("Grayscale")
plt.show()
```

Menampilkan Histogram Gambar Grayscale

Membuat variabel untuk menyimpan data gambar

```python
# digunakan untuk membuat array Numpy dengan nilai awal nol
hg = np.zeros((256))
```

Mengisi setiap nilai dalam array hg dengan 0

```python
for x in range(0, 256):
    hg[x] = 0
```

Menghitung nilai dari gambar

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        hg[gray] += 1
```

Menampilkan Histogram

```python
# plt.figure(figsize=(20, 6))
# plt.plot(hg, color="black", linewidth=2.0)
# plt.show()

# digunakan untuk membuat histogram dari data grayscale yang telah dihitung sebelumnya dan disimpan dalam variabel hg
bins = np.linspace(0, 256, 100)
plt.hist(hg, bins, color="black", alpha=0.5)
plt.title("Histogram")
plt.show()
```

Menampilkan Histogram Kumulatif

```python
# digunakan untuk membuat dua array Numpy baru dengan panjang 256 yang diisi dengan nilai awal nol
hgk = np.zeros((256))
c = np.zeros((256))

for x in range(0, 256):
    hgk[x] = 0
    c[x] = 0

for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        hgk[gray] += 1
                
c[0] = hgk[0]
for x in range(1, 256):
     c[x] = c[x-1] + hgk[x]

hmaxk = c[255]

for x in range(0, 256):
    c[x] = 190 * c[x] / hmaxk

# digunakan untuk membuat histogram kumulatif dari data nilai intensitas 
# piksel yang telah dihitung sebelumnya dan disimpan pada array c
plt.hist(c, bins, color="black", alpha=0.5)
plt.title("Histogram Grayscale Kumulatif")
plt.show()
```

Menampilkan Histogram Hequalisasi

```python
hgh = np.zeros((256))
h = np.zeros((256))
c = np.zeros((256))

for x in range(0, 256):
    hgh[x] = 0
    h[x] = 0
    c[x] = 0

for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        hgh[gray] += 1
                
h[0] = hgh[0]
for x in range(1, 256):
     h[x] = h[x-1] + hgh[x]

for x in range(0, 256):
     h[x] = h[x] / img_height / img_width

for x in range(0, 256):
    hgh[x] = 0
    
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]
        gray = h[gray] * 255
        hgh[int(gray)] += 1

c[0] = hgh[0]
for x in range(1, 256):
     c[x] = c[x-1] + hgh[x]

hmaxk = c[255]

for x in range(0, 256):
    c[x] = 190 * c[x] / hmaxk

#digunakan untuk membuat histogram dari data nilai intensitas 
# piksel yang telah disesuaikan menggunakan teknik histogram equalization
plt.hist(c, bins, color="black", alpha=0.5)
plt.title("Histogram Grayscale Hequalisasi")
# digunakan untuk menampilkan jendela gambar yang sudah dibuat
plt.show()
```
