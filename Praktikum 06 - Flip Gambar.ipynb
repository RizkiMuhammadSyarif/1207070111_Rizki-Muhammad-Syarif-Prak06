Flip Gambar

Import dependency

```python
# memanggil library 
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

Membaca gambar

```python
# digunakan untuk memuat sebuah file gambar bernama "Shikamaru.jpeg" 
# ke dalam sebuah variabel img menggunakan fungsi imread() dari pustaka imageio
img = imageio.imread("Shikamaru.jpeg")
```

Mendapatkan resolusi dan type dari gambar

```python
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype
```

Membuat variabel dengan resolusi dan tipe yang sama seperti gambar

```python
img_flip_horizontal = np.zeros(img.shape, img_type)
img_flip_vertical = np.zeros(img.shape, img_type)
```

Membalik gambar secara horizontalÂ¶

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c]
```

Membalik gambar secara vertical

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        for c in range(0, img_channel):
            img_flip_vertical[y][x][c] = img[img_height-1-y][x][c]
```

Menampilkan hasil balik gambar

```python
# digunakan untuk menampilkan gambar yang sudah diproses flip secara horizontal pada jendela gambar menggunakan pustaka Matplotlib
plt.imshow(img_flip_horizontal)
plt.title("Flip Horizontal")
#digunakan untuk menampilkan jendela gambar yang sudah dibuat
plt.show()

#digunakan untuk menampilkan gambar yang sudah diproses flip secara vertikal pada jendela gambar menggunakan pustaka Matplotlib
plt.imshow(img_flip_vertical)
plt.title("Flip Vertical")
plt.show()
```
